```python
import numpy as np
import pandas as pd
import scipy.stats as st
import skfeature.function as ff
import sklearn.feature_selection as fs
import sklearn.metrics as mt
import sklearn.model_selection as ms
import sklearn.preprocessing as pp
import time as tm
import xgboost as xgb
from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline
```

```python
RANDOM_STATE=4+8+15+16+23+42
NUM_FEATURES_TO_SELECT=10
```

```python
# Cargar datos
X_raw = pd.read_csv("/content/drive/MyDrive/Documentos/7. educación/uniovi/master/tfm/data/BinaryA_features.csv")
y_raw = pd.read_csv("/content/drive/MyDrive/Documentos/7. educación/uniovi/master/tfm/data/BinaryA_labels.csv")
y = y_raw.iloc[:, 1:].values.ravel() # Las etiquetas están en la segunda columna
```

```python
# Guardar los nombres de las características
original_feature_names = X_raw.columns.tolist()
```

```python
# Preprocesamiento
pipe = Pipeline([
  ("feature_elimination", fs.VarianceThreshold()),
  ("scaler", pp.StandardScaler())
])
X_preprocessed = pipe.fit_transform(X_raw)
preprocessed_feature_names = pipe.get_feature_names_out(input_features=original_feature_names)
X = pd.DataFrame(X_preprocessed, columns=preprocessed_feature_names)
```

```python
from skfeature.function.information_theoretical_based.MRMR import mrmr

# Tomar el tiempo inicial
start_time = tm.time()

# Seleccionar las características más relevantes
selected_feature_indices = mrmr(X.values, y, n_selected_features=NUM_FEATURES_TO_SELECT)
selected_feature_indices = np.array(selected_feature_indices).astype(int)
X_selected = X.iloc[:, selected_feature_indices]
selected_features_names = X_selected.columns.tolist()

# Tomar el tiempo final
end_time = tm.time()

# Mostrar las características seleccionadas
print(f"*** Tiempo: {(end_time - start_time):.2f} segundos")
print(
  f"*** Características totales/seleccionadas: "
  f"{len(original_feature_names)}/{len(selected_features_names)}"
)
print(f"")
for name in selected_features_names:
  print(f"{name}")
print("-" * 60)
```
